7. Calculate top-10 people to whom you wrote the biggest amount of messages (name, amount of messages).

```python
toWhomIWrote = df[df['from_id'] == 'PeerUser(user_id=334960821)'].groupby(["to_id"])["id"].count().to_frame().sort_values(by=['id'], ascending=False).rename(columns={"id": "num_of_written_messages_by_me"}).head(10).reset_index()

arr = df_meta['users'].to_numpy()

def idToName(id):
    filter_arr = []

    for user in arr:
        if str(user.split("'user_id': ")[1].split(',')[0]) == str(id):
            filter_arr.append(True)
        else:
            filter_arr.append(False)
    
    users = arr[filter_arr]
    
    if len(users):
        return str('@' + users[0].split("'username': '")[1].split("',")[0])
    else:
        return id


        
toWhomIWrote[['to_id']] = toWhomIWrote[['to_id']].apply(lambda x: [idToName(y) for y in x.to_numpy()])

toWhomIWrote
```

8. Calculate top-10 people who wrote the biggest amount of messages to you (name, amount of messages).

```python
whoWroteMe = df[df['to_id'] == '334960821'].groupby(["dialog_id"])["id"].count().to_frame().sort_values(by=['id'], ascending=False).rename(columns={"id": "num_of_written_messages_to_me"}).head(10).reset_index()

whoWroteMe[['dialog_id']] = whoWroteMe[['dialog_id']].apply(lambda x: [idToName(y) for y in x.to_numpy()])

whoWroteMe
```
